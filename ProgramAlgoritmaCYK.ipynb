```python
# import library yang akan digunakan
import numpy as np
import pandas as pd
```

```python
# pendeklarasian inputan string dan grammar dalam bentuk CNF
inputString = "aabbcc"

grammar =[
    {"S": ("A", "B")},
    {"A": ("C", "D")},
    {"A": ("C", "F")},
    {"B": ("E", "B")},
    {"B": "c"},
    {"C": "a"},
    {"D": "b"},
    {"E": "c"},
    {"F": ("A","D")}
    ]
start = ["S"]
```

```python
# menyatakan panjang string yang akan diperiksa
n = len(inputString)
```

```python
# pengidentifikasian grammar CNF
final = []
nfinal = []

for gr in grammar:
    [[left, right]] = gr.items()
    if len(right) == 1:
        final.append(gr)
    else:
        nfinal.append(gr)
```

```python
# mendeklarasikan array kosong untuk baris dan kolom
P = [[[] for row in range(n)] for x in range(n)]
```

```python
# mengisi tabel baris pertama kolom pertama sampai n
for i in range(n):
    for j in range(len(final)): 
        [[left, right]] = final[j].items()
        if inputString[i] in right:
            P[0][i].append(left)
```

```python
# mengisi kolom pertama sampai (n-baris+1) pada suatu baris
for i in range(1, n):
    for j in range(n-i):
        for k in range(i):
            for nf in nfinal:
                [[left, right]] = nf.items()
                if right[0] in P[k][j] and right[1] in P[i-k-1][j+k+1]:
                    P[i][j].append(left)
```

```python
# memeriksa bagian yang menjadi anggota dan mencetak hasilnya berhasil diterima atau tidak
result = False
for i in P[n-1][0]:
    if i in start:
        result = True
        break;

if result:
    print("Berhasil, String " + inputString + " Dapat di terima")
else:
    print("gagal, String " + inputString + " Tidak dapat di terima")

df = pd.DataFrame(P)
df
```
