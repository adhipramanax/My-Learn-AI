```python
#Nama : Panji Adhipramana 
#NIM : 19051204091
#Kelas : TI 2019 B


#memanggil library math, tempat fungsi sqrt() atau squre root 
import math
# 20x + y - 2z = 17
# 3x + 20y -z = -18
# 2x - 3y + 20z = 25

# memdefinisikan persamaan yang akan diselesaikan setelah diubah menjadi x,y,z
# fungsi lambda mambantu mencari nilai parameter x,y,z sesuai dengan kondisi atau ekspresi setealah tanda titik dua
f1 = lambda x,y,z: (17-y+2*z)/20 # f1 = 0.85000
f2 = lambda x,y,z: (-18-3*x+z)/20 # f2 = 0.90000
f3 = lambda x,y,z: (25-2*x+3*y)/20 # f3 = 1.255000

# mendeklarasikan variabel awal x,y,z sama dengan 0 dan iterasi = 1
x_awal = 0
y_awal = 0
z_awal = 0
iterasi = 1

# batas iterasi adalah jumlah atau bilangan yang menjadi patokan sampai kapan iterasi akan berjalan.
# jika hasil akar dari jumlah x,y,z saat iterasi memiliki nilai yang lebih kecil dari nilai batas iterasi,
# maka iterasi akan berhenti.
e = float(input('Masukkan batas iterasi: '))

# head tabel
print('\nIterasi\tx\ty\tz\t total root awal - total root baru\n')

#mendeklarasikan variabel kondisi untuk parameter loop iterasi
kondisi = True

#iterasi jacobi
while kondisi:
    
    total_awal = x_awal + y_awal + z_awal
    root_awal = math.sqrt(total_awal)
    
    #mencari x baru dengan menggunakan fungsi f1 
    x_baru = f1(x_awal,y_awal,z_awal)
    #mencari y baru dengan menggunakan fungsi f2
    y_baru = f2(x_awal,y_awal,z_awal)
    #mencari z baru dengan menggunakan fungsi f3 
    z_baru = f3(x_awal,y_awal,z_awal)
    
    # mendeklarasikan variabel untuk tempat hasil jumlah dari variabel x,y,z bru
    total_baru = x_baru + y_baru + z_baru
    # mencari akar dari hasil jumlah nilai baru dan ditempatkan pada variabel root_baru
    root_baru = math.sqrt(total_baru)
    
    #mencari norm tak hingg
    e_baru = abs(total_awal - total_baru);
    
    # format float lama print hasil iterasi ke-n dengan 4 angka desimal di belakang titik
    # print('%d\t%0.4f\t%0.4f\t%0.4f\n' %(count, x1,y1,z1)) 
    
    # format print float baru hasil iterasi ke-n dengan 4 angka desimal di belakang titik
    print(str(iterasi)+'\t'+ "{:0.4f}".format(x_baru)+'\t'+"{:0.4f}".format(y_baru)+'\t'+"{:0.4f}".format(z_baru)
            +'\t\t'+ "{:0.5f}".format(e_baru))   
    
    # menambah nilai iterasi dengan angka 1 atau increment
    iterasi += 1
    
    # mendeklarsikan ulang variabel x,y,z awal dengan variable x,y,z, baru
    x_awal = x_baru
    y_awal = y_baru
    z_awal = z_baru
    
    # deklarasi ulang dari variabel kondisi untuk mengakhiri iterasi atau loop
    # iterasi akan berakhir jika kondisi e_baru > e menghasilkan nilai FALSE
    kondisi = e_baru > e

# mencetak nilai akhir dari variabel x,y,z setelah iterasi berakhir
print('\nHasil: x=%0.3f, y=%0.3f dan z = %0.3f\n'% (x_baru,y_baru,z_baru))
```

```python
#memanggil library math, tempat fungsi sqrt() atau squre root 
import math
#memanggil library sebagai nampi untuk membuat array 
#array untuk menampung input user
import numpy as nampi


#menentukan jumlah variable dalam persamaan
# n = int(input("jumlah persamaan dan variabel : "))
#saya tentukan 3 karena saya masih kurang paham iterasi pada deklarasi lambda nanti
n = 3

#membuat array untuk tempat input user dengan ordo(3,4)
equation = nampi.zeros((n,n+1))

#input konstanta variable n ke dalam persamaan
print()
print("Masukkan elemen untuk Matriks A : ")
#membuat iterasi untuk input ke dalam array equation
for x in range(n):
    print("Persamaan ke-"+str(x+1))
    for y in range(n+1):
        #jika y == n maka field tersebut merupakan augment persamaan x
        if y == n:
            equation[x][y] = float(input('Augment persamaan ['+str(x+1)+'] = '))
        else:
        #jika tidak maka field tersebut merupakan konstanta variable
            equation[x][y] = float(input("konstanta ke-"+str(y+1)+" persamaan "+str(x+1)+" = "))

#menampilkan persamaan sesuai dengan input user
#ini untuk memastikan persamaan sesuai dengan yang dikehendaki user
#disini saya menggunakan X1 X2 X3 
print()
print("persamaan anda :")
for x in range(n):
    #membuat variable increment untuk tanda variable ke-n
    increment = 1;
    for y in range(n+1):
        #mencetak sama dengan('=') dilanjutkan isi fieldnya jika y == n
        if y == n:
            print(" = "+str(int(equation[x][y])))
        else:
            #jika isi field sama dengan 0 maka tidak akan dicetak
            if(equation[x][y] == 0):
                continue
            #jika isi field sama dengan 1 maka konstanta tidak perlu dicetak
            elif(equation[x][y] == 1):
                if(y == 0):
                    print("X"+str(increment), end ='')
                else:
                    print(" + X"+str(increment), end ='')
    
                increment = increment + 1
            #jika isi field sama dengan -1 maka konstant juga tidak perlu dicetak namun variable dicetak dengan minus
            elif(equation[x][y]==-1):
                if(y == 0):
                    print("-X"+str(increment), end='')
                else:
                    print(" - X"+str(increment), end='')
                    
                increment = increment + 1
            #jika tidak memenuhi kriteria diatas isi field akan dicetak sesuai kondisi dibawah
            else:
                if(y == 0):
                    print(str(int(equation[x][y]))+"X"+str(increment), end='')
                else:
                    if(equation[x][y]<0):
                        print(" - "+str(int(equation[x][y] * -1))+"X"+str(increment), end='')
                    else:
                        print(" + "+str(int(equation[x][y]))+"X"+str(increment), end='')
                
                increment = increment + 1
                
print()
   
# 4x-y+z = 7
# 4x-8y+z = -21
# -2x+y+5z = 15
#lalu persamaan diatas kita ubah menjadi bentuk seperti dibawah

# x = (7+y-z)/4
# y = (21+4x+z)/8
# z = (15+2x-y)/5

# memdefinisikan persamaan yang akan diselesaikan setelah diubah menjadi x,y,z
# fungsi lambda mambantu mencari nilai parameter x,y,z sesuai dengan kondisi atau ekspresi setealah tanda titik dua
# f1 = lambda x,y,z: (7+y-z)/4
# f2 = lambda x,y,z: (21+4*x+z)/8
# f3 = lambda x,y,z: (15+2*x-y)/5

#mendefinisikan persamaan ke dalam fungsi lambda untuk mencari parameter sesuai kondisi setelah tanda titik
#augment persamaan diletakan di depan lalu ditambah nilai konstanta dikali variable dan -1 karena pindah tempat
#kemudia dibagi dengan konstanta diagonal matrik
f1 = lambda x,y,z: (equation[0][3] + (equation[0][1] * y * -1) + (equation[0][2] * z * -1)) / equation[0][0]
f2 = lambda x,y,z: (equation[1][3] + (equation[1][0] * x * -1) + (equation[1][2] * z * -1)) / equation[1][1]
f3 = lambda x,y,z: (equation[2][3] + (equation[2][0] * x * -1) + (equation[2][1] * y * -1)) / equation[2][2]

# mendeklarasikan variabel awal x,y,z sama dengan 0 dan iterasi = 1
x1_awal = 0
x2_awal = 0
x3_awal = 0
iterasi = 1

# batas iterasi adalah jumlah atau bilangan yang menjadi patokan sampai kapan iterasi akan berjalan.
# jika hasil akar dari jumlah x,y,z saat iterasi memiliki nilai yang lebih kecil dari nilai batas iterasi,
# maka iterasi akan berhenti.
e = float(input('Masukkan batas iterasi: '))

# head tabel
print('\nIterasi\tx\ty\tz\t total root awal - total root baru\n')

#mendeklarasikan variabel kondisi untuk parameter loop iterasi
kondisi = True

#iterasi jacobi
while kondisi:
    
    # mendeklarasikan variabel untuk tempat hasil jumlah dari variabel x,y,z awal
    total_awal = x1_awal + x2_awal + x3_awal
    # mencari akar dari hasil jumlah nilai awal dan ditempatkan pada variabel root_awal
    root_awal = math.sqrt(total_awal)
    
    #mencari x baru dengan menggunakan fungsi f1 
    x1_baru = f1(x1_awal,x2_awal,x3_awal)
    #mencari y baru dengan menggunakan fungsi f2
    x2_baru = f2(x1_awal,x2_awal,x3_awal)
    #mencari z baru dengan menggunakan fungsi f3 
    x3_baru = f3(x1_awal,x2_awal,x3_awal)
    
    # mendeklarasikan variabel untuk tempat hasil jumlah dari variabel x,y,z bru
    total_baru = x1_baru + x2_baru + x3_baru
    # mencari akar dari hasil jumlah nilai baru dan ditempatkan pada variabel root_baru
    root_baru = math.sqrt(total_baru)
    
    # mencari hasil kurang dari hasil jumlah dari variabel awal dan baru
    # menggunakan fungsi abs() atau absolut untuk mendapatkan hasil bilangan positif
    e_baru = abs(total_awal - total_baru);
    
    # format print float lama hasil iterasi ke-n dengan 4 angka desimal di belakang titik
    # print('%d\t%0.4f\t%0.4f\t%0.4f\n' %(count, x1,y1,z1)) 
    
    # format print float baru hasil iterasi ke-n dengan 4 angka desimal di belakang titik
    print(str(iterasi)+'\t'+ "{:0.4f}".format(x1_baru)+'\t'+"{:0.4f}".format(x2_baru)+'\t'+"{:0.4f}".format(x3_baru)
            +'\t\t'+ "{:0.5f}".format(e_baru))    
    
    # menambah nilai iterasi dengan angka 1 atau increment
    iterasi += 1
    
    # mendeklarsikan ulang variabel x,y,z awal dengan variable x,y,z, baru
    x1_awal = x1_baru
    x2_awal = x2_baru
    x3_awal = x3_baru
    
    # deklarasi ulang dari variabel kondisi untuk mengakhiri iterasi atau loop
    # iterasi akan berakhir jika kondisi e_baru > e menghasilkan nilai FALSE
    kondisi = e_baru > e

# mencetak nilai akhir dari variabel x,y,z setelah iterasi berakhir
print('\nHasil: x=%0.3f, y=%0.3f dan z = %0.3f\n'% (x1_baru,x2_baru,x3_baru))

# 20x + y - 2z = 17
# 3x + 20y -z = -18
# 2x - 3y + 20z = 25
```
